#### 扩容数组
数组的大小一般在初始化时候就已经确定，复杂系统中无法爆保证数组之后的内存空间是可用的，从而在大部分编程语言中，数组的长度是不可变的

如果我们希望扩容数组，则需重新建立一个更大的数组，然后把原数组元素依次复制到新数组。这是一个$O(n)$的操作，在数组很大的情况下非常耗时

```python
def extend(nums: list[int], enlarge: int) -> list[int]:
    res = [0] * (len(nums) + enlarge)
    for i in range(len(nums)):
        res[i] = nums[i]
    return res
```

### 数组的优点与局限性
#### 优点：
- 空间效率高：数组为数据分配了连续的内存块，无须额外的结构开销
- 支持随机访问：数组允许在$O(n)$时间内访问任何元素
- 缓存局部性：当访问数组元素时，计算机不仅会加载它，还会缓存其周围的其他数据，从而借助高速缓存来提升后续操作的执行速度

这里的缓存局部性是指：在真正的计算机硬件中，CPU的运算速度远远超过了主内存（RAM）的读取速度。如果CPU每次都要干等着内存把数据送过来，它大部分时间都在“摸鱼”等待。为了解决这个速度差，CPU内部集成了非常昂贵且极速的存储模块，这就是高速缓存（L1 / L2 / L3 Cache）。一次访问会缓存相对应的内存地址附近的一小块数据，将其存储在CPU中

#### 缺点：
- 插入与删除出效率效率低：当数组中的元素较多时，插入和删除操作需要移动大量的元素
- 长度不可变：数组在初始化后长度就固定了，扩容数组需要将所有数据复制到新数组，开销很大。
- 空间浪费：如果数组分配的大小超过了实际所需，那么多余的空间就被浪费了

## 链表
当数组非常大时，内存空间可能无法提供如此大的连续空间。此时采取链表的灵活性就体现出来了。
链表时一种线性数据结构，其中每一个元素都是一个节点对象，各个节点通过引用相连接。引用记录了下一个节点的内存地址，通过它可以从当前节点访问下一个节点。
这时，数据之间的内存地址无需连续。

尾节点在`Python`中被记为`None`

在相同数据量下，链表占据的内存空间比数组大。

```python
class ListNode:
    def __init__(self, val: int):
        self.val : int = val
        self.next : ListNode | None = None
```

### 初始化链表
分为两步，第一步是初始化各个节点对象，第二步是构建节点之间的引用关系。

```python
n0 = ListNode(1)
n1 = ListNode(2)
n2 = ListNode(0)

n0.next = n1 #这一步仅仅是将n1这个节点的内存地址赋值给了n0.next
n1.next = n2
```

通常将头节点作为链表的代称，上述代码中的链表可以记为链表`n0`

### 插入与删除节点
插入只需要改变两个节点的指针即可，时间复杂度为`O(1)`
删除只需要改变一个节点的指针，时间复杂度为`O(1)`；尽管在删除操作之后，被删除节点仍然可以指向下一个节点，但是遍历该链表无法访问被删除节点

```python
def insert(n0: ListNode, P: ListNode):
    n1 = n0.next
    P.next = n1
    n0.next = P

def remove(n0: ListNode):
    # 删除n0后面的节点
    if not n0.next:
        #如果是n0是尾节点
        return
    p = n0.next
    n1 = p.next
    n0.next = n1
```

### 访问与查找节点
在链表中范文节点的效率较低。链表访问要从头开始遍历整个链表。时间复杂度为$O(n)$

```python
def access(head: ListNode, index: int) -> ListNode | None:
    for _ in range(index):
        if not head:
            #这里是防止head为None
            return None
        head = head.next

def find(head: ListNode, target: int) -> int:
    # 遍历链表，查找其中值为target的节点，输出该节点在链表中的索引
    index = 0
    while head:
        if head.val == target:
            return index
        head = head.next
        index += 1
    return -1
```

### 常见链表类型
- 单向链表：即之前介绍的链表。节点包含值和指向下一节点的引用两项。首个节点称为头节点，最后一个节点称为尾节点，尾节点指向`None`.
- 环形列表：将单向链表的尾节点指向头节点，在环形列表中任意节点是为头节点
- 双向链表：双向链表记录了两个方向的引用。定义中同时包括指向后继节点和前驱节点的指针。

```python
class ListNode:
    def __init__(self, val: int):
        self.val = val
        self.next = None
        self.prev = None
```

## 列表
列表是一个抽象的数据结构概念，它表示元素的有序集合，支持元素访问、修改、添加、删除和遍历等操作，无须使用者考虑容量限制的问题。列表可以基于链表或数组实现。

当使用数组实现列表时，长度不可变的性质会导致列表的实用性降低。为解决此问题，我们可以使用***动态数组***来实现列表。它继承了数组的各项优点，并且可以在程序运行过程中进行*动态扩容*。接下来的讨论中，将列表和动态数组等价。

### python中列表的使用
python中的列表`[]`是用动态数组来实现的

```python
# 初始化列表
nums1 = []
nums2 = [1,3,2,3,4]

# 访问元素，时间复杂度为O(1)
num = nums[1]
nums[1] = 0 # 更新元素

# 插入与删除元素
# 在尾部添加元素的时间复杂度为O(1)，插入和删除的时间复杂度与数组相同，时间复杂度为O(n)
# 清空列表
nums.clear()
nums.append(1)
nums.append(2)
nums.append(3)
nums.append(4)

nums.insert(3, 6) # 在索引3处插入数字6
nums.pop(3) # 删除索引3处的元组

# 遍历列表
count = 0
for num in nums:
    count += num

# 拼接列表
nums1 = [9, 8, 3, 5, 5]
nums += nums1 # 将nums1拼接到nums之后

# 排序列表
nums.sort() # 排序之后，列表元素从小到大排列
```

#### 列表的简单实现
设计重点：

- 数量记录：声明一个变量`size`，用于记录当前元素数量，随着元素插入和删除实时更新。更具此变量，我们可以定位列表尾部，以及判断是否需要扩容。
- 扩容机制：若插入元素时列表容量已满，则需要进行扩容。根据扩容倍数创建一个更大的数组，在将当前数组的元素依次移动至新数组。本例中将数组扩容至之前的两倍

数组具有更高的**缓存命中率**，因此它在操作效率上通常优于链表。高缓存效率并不意味着数组在所有情况下都优于链表。

## 关于python中变量赋值以及内存地址的疑问
`python`中`int float str tuple`是不可变对象，`list dict`以及自定义类都是可变对象。
- 不可变对象：对象一旦在内存中被创建，它里面的内容就绝对不允许被改变。任何试图修改该对象内容的操作，本质上都是在内存中开辟一块新的空间，创建一个全新的对象，并把变量名（便利贴）贴到新对象上。

```python
# 代码验证
text = "Hello"
print(id(text))
text1 = text
print(id(text1) == id(text))
# 我们试图给字符串后面加上 " World"
text = text + " World"
print(id(text))
```

- 可变对象：对象在内存中被创建后，它内部包含的数据可以随时被增加、删除或修改，并且这个修改是在原地址上（In-place）直接进行的。对象的内存地址（大箱子的位置）从头到尾都不会变。

```python
# 代码验证
class Province:
    def __init__(self, name, dev_level):
        self.name = name
        self.dev_level = dev_level

    def __repr__(self):
        return f"[{self.name} | 发展度: {self.dev_level}]"

# ================= 实验开始 =================

core_province = Province("法兰西岛", 30)
print(f"步骤 1: core_province 的内存地址是:     {id(core_province)}")

active_selection = core_province
print(f"步骤 2: active_selection 的内存地址是:  {id(active_selection)}")

active_selection.dev_level += 10
print("\n--- 执行发展度 +10 操作 ---")

print(f"步骤 4: 修改后 active_selection 的地址: {id(active_selection)}，数据: {active_selection}")
print(f"步骤 4: 修改后 core_province 的地址:    {id(core_province)}，数据: {core_province}")
```

### 1. C++：物理盒子模型
在 C++ 中，变量等于内存本身。当你声明一个变量时，你是在向计算机的大地（内存）里挖一个固定大小的坑，并给这个坑立了一块永远拔不出来的墓碑（变量名）。

- 声明与绑定： int a = 10;
系统挖了一个 4 字节的坑（比如地址是 0x01），把这个坑命名为 a，然后往里面扔进数字 10。
- 修改值： a = 20;
系统跑回 0x01 这个坑，把里面的 10 倒出来，把 20 扔进去。坑的位置（内存地址）绝对不会变。

- 变量赋值变量： int b = a;
系统在旁边新挖了一个坑叫 b（地址 0x02），然后把 a 坑里的数据复印了一份，扔进 b 坑里。此时 a 和 b 是两个完全独立的物理空间，互不干扰。
### 2. python：便利签模型
变量之间的赋值传递的依然是地址，只不过在涉及修改地址的数据时，依据类型的不同来选择是创建新的内存地址存储新数据还是在原有内存地址的数据基础之上修改
- 创建与绑定： a = 10
系统在内存里吹起了一个装有 10 的气球（对象，地址 0x99），然后拿出一张写着 a 的便利贴，贴在这个气球上。

- 修改值（不可变对象）： a = 20
系统绝对不允许你修改原气球里的 10。它会在旁边重新吹一个装有 20 的新气球（地址 0x88），然后把 a 这个便利贴从旧气球上撕下来，转移贴到新气球上（内存地址发生了改变）。

- 变量赋值变量： b = a
系统根本不会去复印数据。它只是拿出一张新的便利贴 b，啪的一下贴在了 a 正在贴着的那个同一个气球上。

### 有关于python数组的内存地址问题
在`C++`或者`Java`中，创建数组会开辟出一整块的内存空间去存储数据。而在`python`数组中，其是开辟出来一块内存空间取存储它所存储的数据的地址（指针），而数据本身则是随机存储在内存中的。

当你想要访问列表索引 n[2] 时：Python 直接通过数学公式（底层数组基地址 + 索引值 × 指针所占字节数），在 $O(1)$ 的时间内精准定位到第 3 个元素的引用（指针）。然后顺藤摸瓜，通过这个引用找到了散落在内存某处的真正对象（比如链表节点 n3）。这就是为什么哪怕节点对象在内存中天女散花般分布，只要“目录”（引用的集合）是连续排好队的，就能实现 $O(1)$ 的极速访问。

```python
# 1. 初始化两个列表
n = [1, 2, 3]
m = [2, 1, 3]

print("=== 实验一：列表中相同数字的内存地址 ===")
print(f"列表 n 中 '1' 的 id: {id(n[0])}")
print(f"列表 m 中 '1' 的 id: {id(m[1])}")
print(f"单独赋值 x=1 的 id:  {id(1)}")

print("\n=== 实验二：小整数缓存范围 (-5 到 256) ===")
a = 256
b = 256
print(f"a=256 的 id: {id(a)}")
print(f"b=256 的 id: {id(b)}")
print(f"结论：a 和 b 是同一个对象吗？ {a is b}")

print("\n=== 实验三：超出缓存范围的整数 (以 257 为例) ===")
# 使用 int() 动态生成，避开 Python 编辑器的自动代码优化
x = int("257")
y = int("257")
print(f"x=257 的 id: {id(x)}")
print(f"y=257 的 id: {id(y)}")
print(f"结论：x 和 y 是同一个对象吗？ {id(x) ==  id(y)}")
```
