## **Linear Structure**
线性结构（Linear Structure）是一种有序数据项的集合，其中每一个数据项都有唯一的前驱和后驱。
不同线性结构的关键区别在于数据项的增减的方式。常见的有栈（stack），队列（queue），双端队列（deque）和列表（list）

### 栈（stack）
栈是一种有次序的数据项集合，在栈中，数据项的加入和移除仅发生在同一段，这一端叫做栈顶（top），另一端叫栈底（base），这种次序通常叫做LIFO(Last in First Out)

```python
# 使用列表进行栈操作的简单演示
class StackIndexMinus:
    # 将列表中的最后一位作为栈顶
    def __init__(self):
        self.items= []

    def is_empty(self):
        return self.items == []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        return self.items.pop() # list.pop()的时间复杂度为O(1)

    def peak(self):
        return self.items[len(self.items)-1]

    def size(self):
        return len(self.items)

```
