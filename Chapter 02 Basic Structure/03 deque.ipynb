双向队列（deque）在两端都可以进行数据的增减，并且标准库`collections`中的`deque`进行这两个操作的世家复杂的是$O(1)$的

```python
# 进行回文词的判别
from collections import deque
def is_palindrome(word):
    char_deque = deque()
    for char in word:
        char_deque.append(char)

    status = True
    for i in range((len(word) // 2)):
        if char_deque.pop() == char_deque.popleft():
            continue
        else:
            status = False
            break

    return status
```

### 双向队列的实现
#### 基于双向链表的实现

```python
class ListNode:
    def __init__(self, val):
        self.val = val
        self.next = None
        self.prev = None

class LinkedListDeque:
    def __init__(self, val):
        self._front: ListNode|None = None
        self._rear: ListNode|None = None
        self._size: int = 0

    def size(self):
        return self.size()

    def is_empty(self):
        return self._size == 0

    def push(self, val, is_front):
        node = ListNode(val)
        if self.is_empty():
            self._front = self._rear  = node
        elif is_front:
            self._front.prev = node
            node.next = self._front
            self._front = node
        else:
            # 将node添加到链表尾部
            self._rear.next = node
            node.prev = self._rear
            self._rear = node

        self._size += 1

    def push_first(self, num):
        self.push(num, True)

    def push_last(self, num):
        self.push(num, False)

    def pop(self, is_front: bool):
        if self.is_empty():
            raise IndexError("双向队列为空")

        if is_front:
            val = self._front.val
            fnext: ListNode|None = self._front.next

            if fnext is not None:
                fnext.prev = None
                self._front.next = None
            self._front = fnext

        else:
            val = self._rear.val
            rprev: ListNode|None = self._rear.prev

            if rprev is not None:
                rprev.next = None
                self._rear.prev = None
            self._rear = rprev

        self._size -= 1
        return val

    def pop_first(self):
        return self.pop(True)

    def pop_last(self):
        return self.pop(False)

    def peek_first(self):
        if self.is_empty():
            raise IndexError("双向列表为空")
        return self._front.val

    def peek_last(self):
        if self.is_empty():
            raise IndexError("双向列表为空")
        return self._rear.val

    def to_array(self):
        node = self._front
        res = [0] * self.size()
        for i in range(self.size()):
            res[i] = node.val
            node = node.next
        return res
```

#### 基于环形数组实现
只需要增加队首入队和队尾出队的方法

```python
class ArrayDeque:
    def __init__(self, capacity):
        self._nums = [0] * capacity
        self._front = 0
        self._size = 0

    def capacity(self):
        return len(self._nums)

    def size(self):
        return self._size

    def is_empty(self):
        return self.size() == 0

    def index(self, i):
        # 计算环形数组索引
        return (i + self.capacity()) % self.capacity()

    def push_first(self, num):
        if self._size == self.capacity():
            raise IndexError("双向队列已满")

        self._front = self.index(self._front - 1)
        self._nums[self._front] = num
        self._size += 1

    def push_last(self, num):
        if self._size == self.capacity():
            raise IndexError("双向队列已满")

        self._nums[self.index(self._front + self._size)] = num
        self._size += 1

    def peek_first(self):
        if self.is_empty():
            raise IndexError("双向队列为空")
        return self._nums[self._front]

    def peek_last(self):
        if self.is_empty():
            raise IndexError("双向队列为空")
        return self._nums[self.index(self._front + self._size - 1)]

    def pop_first(self):
        num = self.peek_first()
        self._front = self.index(self._front + 1)
        self._size -= 1
        return num

    def pop_last(self):
        num = self.peek_last()
        self._size -= 1
        return num

    def to_array(self):
        if self.is_empty():
            raise IndexError("双向队列为空")
        res = [0] * self._size
        index = self._front
        for i in range(self._size):
            res[i] = self._nums[index]
            index = self.index(index + 1)
        return res
```
