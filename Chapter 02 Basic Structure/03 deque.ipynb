双向队列（deque）在两端都可以进行数据的增减，并且标准库`collections`中的`deque`进行这两个操作的世家复杂的是$O(1)$的

```python
# 进行回文词的判别
from collections import deque
def is_palindrome(word):
    char_deque = deque()
    for char in word:
        char_deque.append(char)

    status = True
    for i in range((len(word) // 2)):
        if char_deque.pop() == char_deque.popleft():
            continue
        else:
            status = False
            break

    return status
```
