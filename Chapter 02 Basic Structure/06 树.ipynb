# 7.1 二叉树
二叉树的基本单元是节点，每个节点包含值、左子节点引用和右子节点引用。

```python
class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left: TreeNode|None = None
        self.right: TreeNode|None = None
```

## 7.1.1 常见二叉树类型
- 完美二叉树（perfect binary tree）：所有层的节点都被完全填满
- 完全二叉树（complete binary tree）：只允许最底层的节点不被完全填满，且最底层的节点必须从左到右依次连续填充（中间不允许有空）
- 完满二叉树（full binary tree）：除了叶节点，其他所有节点都哟两个子节点
- 平衡二叉树（balanced binary tree）：任意节点的左子树和右子树的高度之差的绝对值不超过1

## 7.1.2 二叉树的退化
在最差情况下，二叉树会退化为链表

# 7.2 二叉树的遍历
## 7.2.1 层序遍历
层序遍历（level‑order traversal）从顶部到底部逐层遍历二叉树，并在每一层按照从左到右的顺序访问节点。
层序遍历本质上属于广度优先遍历（breadth‑first traversal），也称广度优先搜索（breadth‑first search,BFS），它体现了一种“一圈一圈向外扩展”的逐层遍历方式。
### (1) 代码实现

```python
from collections import deque
def level_order(root: TreeNode|None) -> list[int]:
    """层序遍历"""
    # 队列遵循先进先出，而广度优先遍历遵循逐层推进，两者相同
    queue: deque[TreeNode] = deque()
    queue.append(root)
    #
    res = []
    while queue:
        node: TreeNode|None = queue.popleft()
        res.append(node.val)

        if node.left is not None:
            queue.append(node.left)
        if node.right is not None:
            queue.append(node.right)
    return res
```

### (2) 复杂度分析
- 时间复杂度：所有节点被访问依次，使用$O(n)$时间，$n$为节点数量
- 空间复杂度：在最差情况下，即使满二叉树，遍历到最底层之前，队列中最多同时存在$\frac{(n+1)}{2}$个节点，占用$O(n)$空间

## 7.2.2 前序、中序、后序遍历
前序、中序、后序遍历属于深度优先遍历（depth-first traversal），也称深度优先搜索（depth-first search,DFS），它体现了一种“先走到尽头，再回溯继续”的遍历方式。每一个节点都会遇到三个位置，分别对应前序遍历、中序遍历以及后序遍历

### (1) 代码实现
深度优先搜索通常基于递归实现

```python
def pre_order(root: TreeNode|None, res: list[int]|None = None):
    if root is None:
        return
    # 访问优先级：根节点 -> 左子树 -> 右子树
    res.append(root.val)
    pre_order(root = root.left)
    pre_order(root = root.right)

def in_order(root: TreeNode|None, res: list[int]|None = None):
    if root is None:
        return
    # 访问优先级：左子树 -> 根节点 -> 右子树
    in_order(root = root.left)
    res.append(root.val)
    in_order(root = root.right)

def right_order(root: TreeNode|None, res: list[int]|None = None):
    if root is None:
        return
    # 访问优先级：左子树 -> 右子树 -> 根节点
    right_order(root.left)
    right_order(root.right)
    res.append(root.val)
```

### (2) 复杂度分析
时间复杂度：所有节点被访问一次，使用$O(n)$时间
空间复杂度：最差情况下，二叉树退化为链表，递归深度达到$n$，系统占用$O(n)$的栈帧空间

# 7.3 二叉树数组表示

## 7.3.1 表示完美二叉树
根据层序遍历的特性，推导出父节点和子节点索引之间的“映射公式”：$若某节点的索引为i，那么该节点的左子节点索引为 2i+1，右子节点索引为2i+2$

## 7.3.2 表示任意二叉树
将二叉树补全为完美二叉树，补全的节点都填充为`None`，在层序遍历中显式表示出所有的`None`

完全二叉树只有最末尾才有`None`的出现，因此可以省略

```python
class ArrayBinaryTree:
    """数组表示下的二叉树类"""
    def __init__(self, arr:list[int|None]):
        self._tree = arr

    def size(self):
        return len(self._tree)

    def val(self, i: int) -> int|None:
        """获取索引为i的节点的值"""
        if i < 0 or i >= self.size():
            return None
        return self._tree[i]

    @staticmethod
    def left(self, i:int) -> int|None:
        """获取索引为i的节点的右子节点的索引"""
        return 2 * i + 1

    @staticmethod
    def right(self, i:int) -> int|None:
        return 2 * i + 2

    @staticmethod
    def parent(self, i:int) -> int|None:
        return (i - 1) // 2

    def level_order(self) -> list[int]:
        """层序遍历"""
        self.res = []
        for i in range(self.size()):
            if self.val(i) is not None:
                self.res.append(self.val(i))
        return self.res

    def dfs(self, i: int, order: str):
        """深度优先遍历"""
        if self.val(i) is None:
            return

        if order == 'pre':
            self.res.append(self.val(i))
        self.dfs(self.left(i), order)

        if order == 'in':
            self.res.append(self.val(i))
        self.dfs(self.right(i), order)

        if order == 'post':
            self.res.append(self.val(i))

    def pre_order(self) -> list[int]:
        self.res = []
        self.dfs(0, order = "pre")
        return self.res

    def in_order(self) -> list[int]:
        self.res = []
        self.dfs(0, order='in')
        return self.res

    def post_order(self) -> list[int]:
        self.res = []
        self.dfs(0, order='post')
        return self.res
```

# 7.4 二叉搜索树
二叉搜索树需要满足以下条件：
1. 对于根节点，左子树中所有节点的值$<$根节点的值$<$右子树中所有节点的值
2. 任意节点的左右子树也是二叉搜索树，即满足条件`1.`

二叉搜索树的查找操作与二分查找算法的工作原理一致，因此时间复杂度为$\mathbf{O}(log n)$


```python
class BinarySearchTree:
    def __init__(self):
        self._root = None

    def search(self, num):
        cur = self._root
        while cur is not None:
            if cur.val > num:
                cur = cur.left
            elif cur.val < num:
                cur = cur.right
            else:
                break
        return cur

    def insert(self, num):
        """
        插入节点，节点插入到叶节点的下方
        类似于搜索，但要保留父节点，用于插入
        插入的时间复杂度是O(log n)
        """
        if self._root is None:
            self._root = TreeNode(num)
            return

        cur, pre = self._root, None
        while cur is not None:
            if cur.val == num:
                # 二叉搜索树不允许两个节点的值重复
                return
            pre = cur
            if cur.val < num:
                cur = cur.right
            else:
                cur = cur.left

        # 插入节点
        node = TreeNode(num)
        if pre.val < num:
            pre.right = node
        else:
            pre.left = node

    def remove(self, num: int):
        """
        删除节点：如果是叶节点直接删除
        如果不是的话，那么则选择该节点的左子树的最大值或者右子树的最小值，将其剪切到删除节点的位置

        删除实际上只是从二叉树中删去了一个数字

        删除的时间复杂度是：O(log n)
        """
        if self._root is None:
            return

        cur, pre = self._root, None

        while cur is not None:
            # 找到待删除结点，跳出循环
            if cur.val == num:
                break

            pre = cur
            if cur.val < num:
                cur = cur.right
            else:
                cur = cur.left
        # 如没找到待删除结点，则直接返回
        if cur is None:
            return

        # 子节点的数量 = 0 or 1
        if cur.left is None or cur.right is None:
            child = cur.left or cur.right
            # 删除cur节点
            if cur != self._root:
                if pre.right == cur:
                    pre.right = child
                else:
                    pre.left = child
        # 子节点的数量 = 2
        else:
            # 使用cur的左子树的最大值所在节点代替cur
            tmp: TreeNode = cur.left
            while tmp.right is not None:
                tmp = tmp.right
            # 使用递归删除二叉树中的tmp节点，该节点并没有从内存中释放
            self.remove(tmp.val)
            cur.val = tmp.val
```

在对二叉搜索树使用中序遍历时，输出的数列时从小到大依次排列的

# 7.5 AVL树
在不断删除以及插入节点之后，原本保持各项操作的时间复杂度为$log n$的二叉树的可能会退化为链表或者左右子树的高度相差极大。
AVL树是用来解决这些的。
AVL树在频繁的增减查改操作之后，仍能保持高效的数据操作性能
## 7.5.1 AVL树常见术语
AVL树既是二叉搜索树，也是平衡二叉树

1. 节点高度

由于AVL树的相关操作需要获取节点高度(指该节点到它最远的叶节点的距离，即所经过的边的数量)，因此需要为节点类添加`height`变量

2. 节点平衡因子
节点的平衡因子定义为节点左子树的高度减去右子树的高度，同时规定空节点的平衡因子为0


```python
class TreeNode:
    def __init__(self):
        self.val = None
        self.height = 0
        self.left = None
        self.right = None
        """叶节点的高度为0，空节点的高度为-1"""

    # 创建两个工具函数用于获取和更新节点的高度
    def height(self, node: TreeNode|None) -> int:
        if node is not None:
            return node.height
        return -1

    def update_height(self, node: TreeNode|None):
        node.height = max([self.height(node.left), self.height(node.right)]) + 1

    def balance_factor(self, node: TreeNode|None) -> int:
        """
        获取平衡因子
        节点的平衡因子定义为节点左子树的高度减去右子树的高度，同时规定空节点的平衡因子为0
        """
        if node is not None:
            return self.height(node.left) - self.height(node.right)
        return 0
        # 一颗AVL树的任意节点的平衡因子index均满足 -1 <= index <= 1
```

## 7.5.2 AVL树旋转
旋转操作既能保持**二叉搜索树**的性质，也能使树重新变为**平衡二叉树**
将平衡因子的绝对值大于1的节点称为失衡节点

| 失衡节点的平衡因子 | 子节点的平衡因子 | 应采取的旋转方式 |
|-----------|----------|----------|
| >1（左偏树）   | >= 0     | 右旋       |
| >1（左偏树）   | < 0      | 先左旋后有右旋  |
| <-1（右偏树）  | <= 0     | 左旋       |
| <-1（右偏树）  | > 0      | 先左旋再右旋   |
