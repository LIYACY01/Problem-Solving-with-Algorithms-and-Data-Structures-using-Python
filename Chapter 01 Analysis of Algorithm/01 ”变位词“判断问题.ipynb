- **问题描述**：所谓变位词是指两个词之间存在组成字母的重新排列关系，如：earth和heart，简单起见，假设参与判断的两个词仅由小写字母组成且长度相等
- **解题目标**：写一个bool函数，以两个词作为参数，返回这两个词是否为变位词

### 解法1：逐词检查，找到之后将相对应的位置的值变为`None`

```python
def anagram_solution1(s1:str, s2:str)->bool:
    list = [*s2]
    pos_1 = 0
    status = True #代表两个词是否为变位词
    while pos_1 < len(s1):
        pos_2 = 0
        found = False
        #对s2中每一个字符进行判断
        while pos_2 < len(s2) and not found:
            if s1[pos_1] == list[pos_2]: # 注意，这里是和list[pos_2]进行比较，而不是s2[pos_2]；后者没有没办法判别这样的两个字符'aab'和'abb'
                found = True
            else:
                pos_2 += 1
        #如果找到，将s2中对应位置赋值变为None，接着找下一个；如果没有找到，那么直接结束循环返回False
        if found:
            list[pos_2] = None
            pos_1 += 1
        else:
            status = False
            break

    return status

anagram_solution1('python', 'typhon')
```

该算法的时间复杂度计算：

对于两个输入长度为$n$的字符串，
`list = [*s2]`的时间复杂度为$O(n)$
`while pos_1 < len(s1):`这个外层循环最坏情况下会执行$n$次（也就是两个词为变位词时）
`while pos_2 < len(s2) and not found:`这个内循环最坏情况下会执行$n$次
所以最坏时间复杂度为$O(n)$

### 解法2：排序比较，将两个字符按照字母顺序排好序，逐字符比较

```python
def anagram_solution2(s1:str, s2:str)->bool:
    list_1, list_2, status = [*s1], [*s2], True
    list_1.sort()
    list_2.sort()

    for i in range(len(list_1)):
        if list_1[i] == list_2[i]:
            continue
        else:
            status = False
            break

    return status

anagram_solution2('cbb', 'ccb')
```

该算法的时间复杂度计算：
注意到：调用的`.sort()`不是没有复杂度的，一般情况下这个函数的复杂度为$O(n \times log(n))$
循环部分的时间复杂度为$O(n)$
所以总的复杂度为$O(n \times log(n))$

```python
# 上面这个算法还有更简洁的写法
def anagram_solution_brief_1(s1:str, s2:str)->bool:
    list_1, list_2 = [*s1], [*s2]
    list_1.sort()
    list_2.sort()
    return list_1 == list_2

anagram_solution_brief_1('ccb', 'cbb')
```

```python
def anagram_solution_brief_2(s1:str, s2:str)->bool:
    return sorted([*s1]) == sorted([*s2])

anagram_solution_brief_2('python', 'typhon')
```

### 解法三：计数比较，为每一个词设计一个26位的计数器，先检查每一个词，记录每个字母出现的次数；之后在对两个词的计数器进行比较

```python
def accumulator(s:str)->list:
    l = [0] * 26
    for i in range(len(s)):
        pos = ord(s[i]) - ord('a')
        l[pos] += 1

    return l

def anagram_solution_3(s1:str, s2:str)->bool:
    return accumulator(s1) == accumulator(s2)

anagram_solution_3('python', 'typhon')
```

该算法的时间复杂度计算：
`accmulator`的时间复杂度为$O(n)$
两个列表比较的时间复杂度为26，也就是$O(1)$
所以该算法的时间复杂度为$O(n)$

```python
# 一个更简介的写法
def accumulator_brief(s:str)->list:
    l = [0] * 26
    for char in s:
        pos = ord(char) - ord('a')
        l[pos] += 1

    return l
```

该算法牺牲了存储空间需要存储两个26节的计数器，换取来时间上的快速

下面试着使用字典实现上面的代码

```python
def accumulator_dict(s:str)->dict:
    char_dict = {}
    for char in s:
        if char in char_dict:
            char_dict[char] += 1
        else:
            char_dict[char] = 1

    return char_dict

def anagram_solution_3_dict(s1:str, s2:str)->bool:
    return accumulator_dict(s1) == accumulator_dict(s2)

anagram_solution_3('python', 'typhon')
```

```python
# 更简洁的一种写法
def accumulator_dict_brief(s:str)->dict:
    char_dict = {}
    for char in s:
        char_dict[char] = char_dict.get(char, 0) + 1

    return char_dict
```
